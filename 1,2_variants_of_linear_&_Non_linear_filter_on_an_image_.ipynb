# **CONTOUR **

```python
from PIL import Image, ImageFilter
from PIL.ImageFilter import (BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EDGE_ENHANCE_MORE,EMBOSS, FIND_EDGES, SMOOTH, SMOOTH_MORE, SHARPEN)

img = Image.open('/content/drive/MyDrive/Colab Notebooks/images/1-2.png')
img1 = img.filter(CONTOUR)
img1
```

```python
# DETAIL
img2 = img.filter(DETAIL)
img2
```

```python
# EDGE_ENHANCE
img3 = img.filter(EDGE_ENHANCE)
img3
```

```python
# EDGE_ENHANCE_MORE
img4=img.filter(EDGE_ENHANCE_MORE)
img4
```

```python
# EMBOSS
img5=img.filter(EMBOSS)
img5
```

```python
# FIND_EDGES
img6 = img.filter(FIND_EDGES)
img6
```

```python
# SMOOTH
img7 = img.filter(SMOOTH)
img7
```

```python
# SHARPEN
img8 = img.filter(SHARPEN)
img8
```
